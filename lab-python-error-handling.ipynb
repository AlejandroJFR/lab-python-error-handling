````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory


def get_customer_orders(inventory, products):
    customer_orders = set()
    try: 
        num_orders = int(input("Enter the number of customer orders: "))
        
        if num_orders < 0:
            raise ValueError("Invalid number of customer orders! Please enter a non-negative number")
        
    except ValueError as e:
        print(f"Error: {e}")
    
    for i in range(num_orders):
        valid_input = False
        while not valid_input:
            try:
                product = input(f"Enter product {i + 1} of your order").lower()
                if product not in products or inventory[product] == 0: 
                    raise ValueError("Item not in inventory")
                else:
                    customer_orders.add(product)
                    valid_input = True 
            except ValueError as e:
                print(f"Error: {e}")
    return customer_orders
        
    
def update_inventory(customer_orders, inventory):
    for item in customer_orders:
        if inventory[item] != 0:
            inventory[item] -= 1
        else:
            print("Item not in inventory: ", item)
            inventory.pop(item)
            
            
def calculate_order_statistics(customer_orders, products):
    return len(customer_orders), (len(customer_orders)/len(products)) * 100


def calculate_total_price(customer_orders):
   prices = {}
   for product in customer_orders:
      valid_input = False
      while not valid_input:
         try:
            price = float(input(f"Please enter the price of {product}"))
            
            if price < 0:
               raise ValueError("Price cannot be a negative value. Try again!")

            prices[product] = price
            valid_input = True
            break
        
         except ValueError as e:
            print(f"Invalid input: {e}. Try again!")
   
   return sum(prices.values())


def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print(f"Total Products Ordered: {order_statistics[0]}")
    print(f"Percentage of Products Ordered: {round(order_statistics[1])}%") 
    
    
def print_inventory(inventory):
    for product, quantity in inventory.items():
            print(f"{product.capitalize()}s: {quantity}")
````

````python
# Define data structures
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Calls to functions
inventory = initialize_inventory(products)
print(inventory)
customer_orders = get_customer_orders(inventory, products)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
update_inventory(customer_orders, inventory)
print_inventory(inventory)
total = calculate_total_price(customer_orders)
print(f"Total Price: ${total}")
````
